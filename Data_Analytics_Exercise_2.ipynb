```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
iris = datasets.load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
print("Iris Dataset Overview:")
print(iris_df.head())
```

1. Data Loading and Overview :

1) Data Loading :

The Iris dataset are loaded using a function called datasets.load_iris() from sklearn, containing features such as sepal length, sepal width, petal length, and petal width, together with target classes. The data are converted into a Pandas DataFrame to ensure ease of manipulation, and the target variable is added as a new column for species classification.

2) Display the Data :

This displays the first few entries in the dataset, giving the structure and content.


```python
random_sample = iris_df.sample(n=30, random_state=42)
print("\nRandom Sample (30 observations):")
print(random_sample)
```

2. Simple Random Sampling :

1) Random Sampling :

Randomly samples 30 observations from the dataset using sample(). random_state=42 ensures the reproducibility of the random sampling.
The sampled data is displayed using:

print("
Random Sample (30 observations):")
print(random_sample)

```python
sample_means = [iris_df.sample(n=30, random_state=i)['sepal length (cm)'].mean() for i in range(100)]
plt.figure(figsize=(8, 6))
plt.hist(sample_means, bins=15, edgecolor='k')
plt.title('Distribution of Sample Means (Sepal Length)')
plt.xlabel('Mean Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

3. Analysis of the Distribution of Sample Means:

1) Calculation of Mean of Multiple Samples:

The random sampling is repeated 100 times, using different random states (i ranges from 0 to 99).
For each sample of size 30, the average of sepal length is calculated.

2) Distribution of the Sample Means Plot :

A histogram depicts the distribution of sample means. As an illustration of the Central Limit Theorem, the averages of random samples represent a normal distribution even when the underlying data is not perfectly normal.

```python
def systematic_sampling(df, fraction):
    n = len(df)
    step = int(1 / fraction)
    indices = np.arange(0, n, step)
    return df.iloc[indices]

systematic_sample = systematic_sampling(iris_df, fraction=0.2)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(iris_df.index, iris_df['sepal length (cm)'], alpha=0.7)
plt.title('Original Dataset')
plt.xlabel('Index')
plt.ylabel('Sepal Length')

plt.subplot(1, 2, 2)
plt.scatter(systematic_sample.index, systematic_sample['sepal length (cm)'], alpha=0.7, color='orange')
plt.title('Systematic Sample (20% of Dataset)')
plt.xlabel('Index')
plt.ylabel('Sepal Length')

plt.tight_layout()
plt.show()
```

4. Systematic Sampling :

1) Function for Systematic Sampling :

Select observations at fixed intervals. fraction specifies the sampling fraction, for instance, 0.2 for 20% of the data. Calculate the step size 1 / fraction. Create indices starting from 0 with fixed interval step. Use iloc to select rows at these indices.

2) Apply Systematic Sampling :

Selects 20% of the dataset systematically.

3) Visualization :

The Left Plot is Scatter plot of sepal length in the original dataset and
the Right Plot is Scatter plot of the systematically sampled data (20% of the original dataset). Visually compares the distribution of the original data and the systematic sample.
